### Now doing it for multiple pages and then making a single dataframe for all pages

```python
final = pd.DataFrame()

for j in range(1,501):
    url = 'https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers=headers, verify=False).text
    
    soup = BeautifulSoup(webpage, 'lxml')
    
    content = soup.find_all('div', class_='companyCardWrapper__metaInformation')
    
    company = []
    business_type = []
    hrf = []
    crf = []


    for i in content:
        company.append(i.find('h2').text.strip())
        business_type.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())
        highly_rated = i.find('span', class_='companyCardWrapper__ratingHeader--high')
        if highly_rated:
            highly_rated_value = highly_rated.find_next('span', class_='companyCardWrapper__ratingValues')
            hrf.append(highly_rated_value.text.strip() if highly_rated_value else None)
        else:
            hrf.append(None)
        critically_rated = i.find('span', class_='companyCardWrapper__ratingHeader--critical')
        if critically_rated:
            critically_rated_value = critically_rated.find_next('span', class_='companyCardWrapper__ratingValues')
            crf.append(critically_rated_value.text.strip() if critically_rated_value else None)
        else:
            crf.append(None)
            
    df_dict = {'company':company, 'business_type':business_type, 'hrf':hrf, 'crf':crf}
    
    df = pd.DataFrame(df_dict)
    df2 = df.business_type.str.split('|', expand=True)
    result = pd.concat([df,df2], axis=1)
    
    # now combining two dataframe
    final = final._append(result)
```
